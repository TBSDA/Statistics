# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.linear_model as slm
```

Define the function.

```python
x = 3 * np.random.rand(200, 1)
```

```python
y = 2 + 2*x + np.random.randn(200, 1)
```

Plot the generated data set.

```python
plt.figure(figsize=(15,8))
plt.title('Randomly generated linear data set', size='x-large')
plt.scatter(x, y)
plt.xlabel('x', loc='right', size='large')
plt.ylabel('y', rotation=0, size='large')
```

Linear regresion by normal equation.

```python
# adding 1 to every sample
x_b = np.c_[np.ones((200, 1)), x]
```

```python
# solving the theta^ = (X.T*X)^-1 * X.T * y
theta = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
```

Check the result. The orginal function is y=2+2x+gaussian noise

```python
theta
```

Check the prediction.

```python
x_test = np.array([[0], [3]])
x_test_b = np.c_[np.ones((2, 1)), x_test]
```

```python
y_pred = x_test_b.dot(theta)
```

```python
y_pred
```

Let's put it on the above chart.

```python
plt.figure(figsize=(15,8))
plt.title('Linear regression for randomly generated linear data set', size='x-large')
plt.scatter(x, y)
plt.plot(x_test, y_pred, color='black', linewidth=3)
plt.xlabel('x', loc='right', size='large')
plt.ylabel('y', rotation=0, size='large')
```

Linear regresion by sklearn.

```python
linear_reg = slm.LinearRegression()
```

```python
linear_reg.fit(x, y)
```

Check the results. 
.coef_ stores the coefficients;
.intercept_ stores the y-intercept.

```python
linear_reg.intercept_, linear_reg.coef_
```

Check the prediction.

```python
y_pred_lm = linear_reg.predict(x_test)
```

Let's put it as well on the above chart.

```python
plt.figure(figsize=(15,8))
plt.title('Linear regression for randomly generated linear data set', size='x-large')
plt.scatter(x, y)
plt.plot(x_test, y_pred_lm, color='black', linewidth=3)
plt.xlabel('x', loc='right', size='large')
plt.ylabel('y', rotation=0, size='large')
```
