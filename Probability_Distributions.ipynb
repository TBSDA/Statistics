# Probability Distributions

Presentation of a couple common probability distributions.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
```

General plot function.

```python
def plot_dist(data, name):
    plt.figure(figsize=(15, 8))
    # sns.displot(p_binom) # it creates separate chart (does not respect the plt.figure) - some bug? (v.0.11.2)
    sns.histplot(data, kde=True)
    plt.title(f'The {name}', size='x-large')
    plt.xlabel(name)
    plt.ylabel('Frequency')
```

All descriptions come from Wikipedia.

1. Binomial Distribution

The binomial distribution with parameters n and p is the discrete probability distribution of the number of successes in a sequence of n independent experiments, each asking a yes–no question, and each with its own Boolean-valued outcome: success (with probability p) or failure (with probability q = 1 − p).

```python
# n - number of trails
# p - probability of success
# n & p are shape params
# size - number of times to repeat the trials
pd_binom = stats.binom.rvs(n = 15, p = 0.75, size=2500)
```

```python
pd_binom
```

```python
plot_dist(pd_binom, 'Binomial Distribution')
```

2. Exponential Distribution

The exponential distribution is the probability distribution of the time between events in a Poisson point process, i.e., a process in which events occur continuously and independently at a constant average rate. It is a particular case of the gamma distribution. 

```python
# scale - shape param; it is a 1/lambda
# loc - shift argument
# size - number of random variates
pd_expon = stats.expon.rvs(scale=1,loc=0,size=2500)
```

```python
plot_dist(pd_expon, 'Exponential Distribution')
```

3. Gamma Distribution

The gamma distribution is a two-parameter family of continuous probability distributions. The maximum entropy probability distribution (both with respect to a uniform base measure and with respect to a 1/x base measure) for a random variable X (please check the conditions for X).

```python
# a - shape param; if 'a' is an int - Erlang distribution, if 'a'=1 - Exponential distribution
# size - number of random variates
pd_gamma = stats.gamma.rvs(a=3, size=2500)
```

```python
plot_dist(pd_gamma, 'Gamma Distribution')
```

4. Normal Distribution

The normal distribution (also known as Gaussian, Gauss, or Laplace–Gauss distribution) is a type of continuous probability distribution for a real-valued random variable. It is a structure used to approximate some phenomenon's long-term behaviour. 

```python
# scale - standard deviation
# loc - mean of the distribution
# size - number of random variates
pd_normal = stats.norm.rvs(scale=1, loc=0, size=2500)
```

```python
plot_dist(pd_normal, 'Normal Distribution')
```

5. Uniform Distribution (continuous)

The continuous uniform distribution (or rectangular distribution) is a family of symmetric probability distributions. The distribution describes an experiment where there is an arbitrary outcome that lies between certain bounds. The bounds are defined by the parameters, a and b, which are the minimum and maximum values. The interval can either be closed (e.g. [a, b]) or open (e.g. (a, b)).

```python
# scale - interval argument 
# loc - location parameter
# loc is a min value and loc+scale is max value
# size - number of random variates
pd_uniform = stats.uniform.rvs(scale=10, loc=6, size=2500)
```

```python
plot_dist(pd_uniform, 'Uniform Distribution')
```

6. Poisson Distribution

The is poisson distribution a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event. The Poisson distribution can also be used for the number of events in other specified interval types such as distance, area or volume.

```python
# mu - lambda
# size - number of random variates
pd_poisson = stats.poisson.rvs(mu=5, size=2500)
```

```python
plot_dist(pd_poisson, 'Poisson Distribution')
```
